üîπ Passo 1: Criando o Dataset
üìå Tarefa:

Gere um dataset com √°rea do im√≥vel (m¬≤) como vari√°vel preditora e pre√ßo do im√≥vel (R$) como vari√°vel alvo.
Inclua um outlier (500m¬≤, R$10.000) e observe como isso afeta o modelo.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Criando um dataset fict√≠cio usando np.array([VALORES]).reshape(-1, 1)
# Crie um vetor y com valores de imoveis np.array([VALORES])
X = ?  # √Årea em m¬≤
y = ?  # Pre√ßo do im√≥vel em R$ mil

# Criando um DataFrame
df = pd.DataFrame({"√Årea (m¬≤)": X.flatten(), "Pre√ßo (R$ mil)": y})

X = df["?"]  # √Årea em m¬≤
y = df["PRICE"]  # Pre√ßo do im√≥vel em R$ mil

# Visualizando os dados
plt.scatter(X, y, color="blue", alpha=0.6)
plt.xlabel("√Årea do Im√≥vel (m¬≤)")
plt.ylabel("Pre√ßo do Im√≥vel (R$ mil)")
plt.title("Dataset de Pre√ßos de Im√≥veis (Com Outlier)")
plt.show()
```

üîπ Passo 2: Treinando o Modelo de Regress√£o Linear
üìå Tarefa:

Treine um modelo de Regress√£o Linear com os dados acima e visualize a reta de regress√£o.

```python
# Criando e treinando o modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Fazendo previs√µes
y_pred = modelo.predict(X)

# Visualizando a reta de regress√£o
plt.scatter(X, y, color="blue", alpha=0.6, label="Dados Reais")
plt.plot(X, y_pred, color="red", label="Regress√£o Linear")
plt.xlabel("√Årea do Im√≥vel (m¬≤)")
plt.ylabel("Pre√ßo do Im√≥vel (R$ mil)")
plt.title("Regress√£o Linear com Outlier")
plt.legend()
plt.show()
```

üîπ Passo: Removendo o Outlier e Comparando

üìå Tarefa:

Remova o outlier (500m¬≤, R$10.000) e treine novamente o modelo.

Compare a nova reta de regress√£o com a anterior.

```python
# Removendo o outlier
X_sem_outlier = X[:-1]  # Removendo o √∫ltimo valor
y_sem_outlier = y[:-1]  # Removendo o √∫ltimo valor

# Treinando o modelo sem o outlier
modelo_sem_outlier = LinearRegression()
modelo_sem_outlier.fit(X_sem_outlier, y_sem_outlier)

# Fazendo previs√µes
y_pred_sem_outlier = modelo_sem_outlier.predict(X_sem_outlier)

# Visualizando as duas regress√µes
plt.scatter(X_sem_outlier, y_sem_outlier, color="blue", alpha=0.6, label="Dados Reais (Sem Outlier)")
plt.plot(X, y_pred, color="red", linestyle="--", label="Regress√£o Com Outlier")  # Modelo anterior
plt.plot(X_sem_outlier, y_pred_sem_outlier, color="green", label="Regress√£o Sem Outlier")  # Novo modelo
plt.xlabel("√Årea do Im√≥vel (m¬≤)")
plt.ylabel("Pre√ßo do Im√≥vel (R$ mil)")
plt.title("Compara√ß√£o: Reg. Linear Com e Sem Outlier")
plt.legend()
plt.show()
```

üîπ Passo 4: Medindo o Impacto no Erro
üìå Tarefa:

Calcule o Erro Quadr√°tico M√©dio (MSE) para comparar os dois modelos.

```python
from sklearn.metrics import mean_squared_error

# Calculando o erro MSE para ambos os modelos
mse_com_outlier = mean_squared_error(y, modelo.predict(X))
mse_sem_outlier = mean_squared_error(y_sem_outlier, modelo_sem_outlier.predict(X_sem_outlier))

print(f"Erro Quadr√°tico M√©dio (MSE) com outlier: {mse_com_outlier:.2f}")
print(f"Erro Quadr√°tico M√©dio (MSE) sem outlier: {mse_sem_outlier:.2f}")
```
